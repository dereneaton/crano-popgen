### Technical replicates
We included 8 technical replicates in this data set. Preliminary analyses here as well as full analyses elsewhere (see Eaton & Ree xx) showed that when assembled these replicate samples assembled nearly identically to one another, and thus for this analysis I combine technical replicates into a single sample with the script below. This reduces the data set to 107 sampled individuals. 

```python
## make a directory to backup originals
! mkdir analysis_pyrad/replicates/

## find which ones are technical replicates
samples = glob.glob("analysis_pyrad/edits/*.edit")
samplenames = [i.split("/")[-1].split("_")[0][1:].replace(".edit","") for i in samples]
replicates = [ i for i in samplenames if samplenames.count(i) >1 ]

## combine r12 and p12 replicate samples into h12
for rep in replicates:
    p1,p2 = glob.glob("analysis_pyrad/edits/*"+rep+".edit")
    print p1,p2
    a,b,c = p1.split("/")
    p3 = "/".join([a,b,'h'+c[1:]])
    wait = ! cat $p1 $p2 > $p3
    wait = ! mv $p1 analysis_pyrad/replicates/
    wait = ! mv $p2 analysis_pyrad/replicates/

## change name of 6.8 and 26.6 to match other inds in same population
## ! mv analysis_pyrad/edits/r6.8.edit analysis_pyrad/edits/d6.8.edit
## ! mv analysis_pyrad/edits/p26.6.edit analysis_pyrad/edits/d26.6.edit
```
